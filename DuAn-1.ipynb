```python
movies = df[['title', 'on_streaming_date', 'genre','certification','runtime','rating',
             'content','Director','cast1','cast2','cast3',
             'Budget','Revenue','keyword'
            ]]
```

```python
movies[['Năm', 'Tháng','Ngày']] = movies['on_streaming_date'].str.split('-', expand=True)
```

```python
movies['on_streaming_date'] = pd.to_datetime(movies['on_streaming_date'])
movies['Thứ'] = movies['on_streaming_date'].dt.weekday
```

```python
def convert_thu_to_chu(Thứ):
    if Thứ == 0:
        return "Thứ 2"
    elif Thứ == 1:
        return "Thứ 3"
    elif Thứ  == 2:
        return "Thứ 4"
    elif Thứ == 3:
        return "Thứ 5"
    elif Thứ == 4:
        return "Thứ 6"
    elif Thứ == 5:
        return "Thứ 7"
    elif Thứ == 6:
        return "Chủ nhật"
    else:
        return "Không hợp lệ"
movies['Thứ'] = movies['Thứ'].apply(convert_thu_to_chu)
```

```python
movies['Năm'] = movies['Năm'].astype(int)
movies['Tháng'] = movies['Tháng'].astype(int)
movies['Ngày'] = movies['Ngày'].astype(int)
```

```python
movies['Profit'] = movies['Revenue'] - movies['Budget']
```

```python


# Set the figure size
plt.figure(figsize=(17,6))

# Create the count plot
ax = sns.countplot(x="Năm", data=movies, color='skyblue')

# Set the title
ax.set_title('Count of Movies by Year', fontsize=25)

# Add numerical values to the bar plot
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.0f'), (p.get_x() + p.get_width() / 2., p.get_height()),
                ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points', fontsize=14)

# Rotate the x-axis tick labels for better readability
plt.xticks(rotation=90,fontsize=10)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create bins for runtime
bin_size = 20
Q1['runtime_bin'] = pd.cut(Q1['runtime'], bins=range(0, Q1['runtime'].max() + bin_size, bin_size))

# Group the data based on the runtime bins
grouped_data = Q1.groupby('runtime_bin').agg({'title': 'count',
                                             'Adjusted_Revenue': 'mean',
                                             'Adjusted_Profit': 'mean'}).reset_index()

# Data for the plot
runtime_bin = grouped_data['runtime_bin'].astype(str)  # Convert to string
film_count = grouped_data['title']
revenue_mean = grouped_data['Adjusted_Revenue']
profit_mean = grouped_data['Adjusted_Profit']

# Increase figure size
plt.figure(figsize=(23, 10))

# Create the axis objects
ax1 = plt.gca()

# Plot the film count as a bar chart
bars = ax1.bar(runtime_bin, film_count, color='#b5eae6')
ax1.set_xlabel('Runtime', fontsize=20)
ax1.set_ylabel('Film Count', color='b', fontsize=20)
ax1.tick_params('y', colors='b')

# Create a twin axis for revenue and profit
ax2 = ax1.twinx()

# Plot the adjusted revenue and profit as line charts
revenue_line = ax2.plot(runtime_bin, revenue_mean, color='r', marker='o', linestyle='-', label='Adjusted Revenue')
profit_line = ax2.plot(runtime_bin, profit_mean, color='purple', marker='o', linestyle='-', label='Adjusted Profit')
ax2.set_ylabel('Adjusted Revenue / Adjusted Profit', color='k', fontsize=20)
ax2.tick_params('y', colors='k')

# Add film count as annotations on top of the bars
for bar, count in zip(bars, film_count):
    height = bar.get_height()
    ax1.annotate(count, xy=(bar.get_x() + bar.get_width() / 2, height), xytext=(0, 3),
                 textcoords="offset points", ha='center', va='bottom', fontsize=15)

# Add numbers to the revenue line
for x, y in zip(runtime_bin, revenue_mean):
    ax2.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=17)

# Add numbers to the profit line
for x, y in zip(runtime_bin, profit_mean):
    ax2.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=17)

# Set the title and legend
plt.title('Film Count, Adjusted Revenue, and Adjusted Profit by Runtime', fontsize=20)
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(fontsize=20,rotation=90)

# Display the plot
plt.show()
```

```python
import random
import matplotlib.pyplot as plt
import pandas as pd

# Ghép các cột cast1, cast2, cast3 thành một cột cast
Q1['cast'] = Q1[['cast1', 'cast2', 'cast3']].stack().reset_index(drop=True)

# Lựa chọn top 10 diễn viên theo số lượng phim
top_actors = Q1['cast'].value_counts().head(20)

# Tính tổng số lượng phim cho mỗi diễn viên
film_count = Q1['cast'].value_counts().sort_values(ascending=False)


# Tính trung bình của Adjusted_Revenue và Adjusted_Profit cho mỗi diễn viên
revenue_mean = Q1.groupby('cast')['Adjusted_Revenue'].mean()
profit_mean = Q1.groupby('cast')['Adjusted_Profit'].mean()

# Tạo màu duy nhất cho cột
bar_color = '#B5EAD7'

# Tạo biểu đồ
plt.figure(figsize=(15, 8))

# Vẽ cột thanh cho số lượng phim
ax1 = plt.gca()
film_count[top_actors.index].plot(kind='bar', color=bar_color, ax=ax1)
ax1.set_xlabel('Diễn viên', fontsize=12)
ax1.set_ylabel('Số lượng phim', fontsize=12)
ax1.set_title('Số lượng phim của mỗi diễn viên (Top 10)', fontsize=14)
ax1.tick_params(axis='x', labelrotation=45)

# Vẽ đường cho Adjusted_Revenue và Adjusted_Profit
ax2 = ax1.twinx()
revenue_mean[top_actors.index].plot(color='r', marker='o', linestyle='-', ax=ax2, label='Adjusted Revenue')
profit_mean[top_actors.index].plot(color='purple', marker='o', linestyle='-', ax=ax2, label='Adjusted Profit')
ax2.set_ylabel('Adjusted Revenue / Adjusted Profit', fontsize=12)
ax2.tick_params(axis='y')

# Hiển thị số trên đường
for x, y in enumerate(revenue_mean[top_actors.index]):
    ax2.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=10)

for x, y in enumerate(profit_mean[top_actors.index]):
    ax2.text(x, y, f'{y:.2f}', ha='center', va='bottom', fontsize=10)

# Hiển thị chú thích
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Hiển thị biểu đồ
plt.tight_layout()
plt.show()
```
