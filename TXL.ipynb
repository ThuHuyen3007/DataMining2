```python
movie = pd.read_csv('Movie.csv')
movie
```

```python
for i in range(len(movie)):
    movie['on_streaming_date'][i] = movie['on_streaming_date'][i][:10]
from datetime import datetime
movie['on_streaming_date'] = pd.to_datetime(movie['on_streaming_date'], format='%m/%d/%Y')
```

```python
d=movie[movie['on_streaming_date'].dt.year>=2000]
d[(d['Budget']!='-') & (d['Revenue']!='-')].shape
```

```python
df=movie[(movie['Budget']!='-') & (movie['Revenue']!='-')]
df['tagline'] = df['tagline'].fillna('')
df['content'] = df['tagline'] + '\n' + df['overview']
df=df[['url', 'title', 'on_streaming_date', 'genre', 'certification', 'runtime', 'rating', 'content', 'Director',
      'cast1', 'cast1_link', 'cast2', 'cast2_link', 'cast3', 'cast3_link', 'Status', 'Original Language', 'Budget', 'Revenue', 'keyword']]
df.info()
```

```python
df=df.dropna(axis=0)
df=df.drop_duplicates()
df = df.reset_index(drop=True)
df
```

```python
for i in range(len(df)):
    df['Budget'][i] = df['Budget'][i][1:]
    df['Revenue'][i] = df['Revenue'][i][1:]
```

```python
from datetime import datetime
df['on_streaming_date'] = pd.to_datetime(df['on_streaming_date'], format='%m/%d/%Y')
df['Budget'] = df['Budget'].str.replace(',', '').astype('float')
df['Revenue'] = df['Revenue'].str.replace(',', '').astype('float')
```

```python
df = df[df['on_streaming_date'].dt.year>=2000]
```

```python
def convert_to_minutes(time_string):
    if len(time_string) < 5:
        hours = '0'
        minutes = time_string
    else:
        hours, minutes = time_string.split('h ')
    minutes = minutes[:-1]  # Loại bỏ ký tự 'm' cuối cùng
    total_minutes = int(hours) * 60 + int(minutes)
    return total_minutes
df['runtime'] = df['runtime'].apply(convert_to_minutes)
```

```python
df.info()
```

```python
df.describe()
```

```python
from scipy import stats
z_scores = stats.zscore(df['runtime'])
df = df[(z_scores < 3) & (z_scores > -3)]
```

```python
df.describe()
```

```python
movie_txl = pd.read_csv('Movie_TXL.csv')
movie_txl = movie_txl[['url','Director_link']]
movie_txl
```

```python
df = pd.merge (left = df, left_on = 'url', right = movie_txl, right_on = 'url')
df.columns
```

```python
df = df[['url', 'title', 'on_streaming_date', 'genre', 'certification',
       'runtime', 'rating', 'content', 'Director', 'Director_link', 'cast1', 'cast1_link',
       'cast2', 'cast2_link', 'cast3', 'cast3_link', 'Status',
       'Original Language', 'Budget', 'Revenue', 'keyword']]
df = df.reset_index(drop=True)
df.to_csv('Movie_TXL.csv', index=False)
df
```
